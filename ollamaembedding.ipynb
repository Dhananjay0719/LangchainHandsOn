```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(OllamaEmbeddings(model="gemma:2b"))    # After Langchain 0.3.1 OllamaEmbeddings() which used llama2 by default was deprecated due to which now model name is compulsory
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(["Alpha is first letter of greek","Beta is second alphabet of greek"])
```
