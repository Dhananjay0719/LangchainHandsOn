VECTOR STORE USING FAISS CHROMA 

FACEBOOK AI SIMILARITY SEARCH (FAISS)

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader=TextLoader("dummy_file.txt")
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=20,chunk_overlap=0)
docs=text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings=(OllamaEmbeddings(model="gemma:2b"))
```

```python
db=FAISS.from_documents(docs,embeddings)
```

```python
db
```

```python
#querying the database
query="What was the dummy file intended for?"
docs=db.similarity_search(query)
```

```python
docs[0]
```

```python
docs[0].page_content
```

```python
#VECTORSTORES ARE CONVERTED IN RETRIEVER CLASS.THINK OF AN INTERFACE CONNECTED TO VECTOR DB WHICH HELPS TO RETREIVE DESIRED THINGS

retriever=db.as_retriever()
docs=retriever.invoke(query)
```

```python
docs[0].page_content
```

```python
#SIMILARITY SEARCH WITH SCORE (L2 SCORE ALSO KNOWN AS MANHATTAN DISTANCE)
docs_and_score=db.similarity_search_with_score(query)
```

```python
docs_and_score
```

```python
db.save_local("faiss_index")
```
